```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("E:\\datasets\\CrimesOnWomenData.csv")
print(df.head().to_string())
```

```python
def lower(x):
    return x.lower()
df.State=df.State.apply(lower)
print(df.head())
```

```python
df.columns=df.columns.str.lower()
print(df.columns)
```

```python
print(df.state.unique())
```

```python
df["total crime"]=df["rape"]+df["k&a"]+df["dd"]+df["aow"]+df["aom"]+df["dv"]+df["wt"]
# print(df.head(10).to_string())
print(df.sort_values(by="total crime").head(20))
```

```python
k=df.groupby("state")["total crime"].sum()
print(k.sort_values())              # laskshadweep has lowest crime and Uttar Pradesh has highest crime between 2001 & 2021
```

```python
gp=df.groupby("state").get_group("uttar pradesh")
print(gp)
```

```python
sns.set_style("dark") 
sns.barplot(x=gp.year,y=gp["total crime"])
plt.xticks(rotation=45)  
plt.tight_layout() 
plt.title("U.P criminal cases (2001-2021)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
k=df[df.state=="uttar pradesh"]
plt.plot(k.year,k.rape,label="RAPE")
plt.plot(k.year,k["k&a"],label="KIDNAPPING")
plt.plot(k.year,k.dd,label="DOWRY DEATh")
plt.plot(k.year,k.aow,label="Assault on Women")
plt.plot(k.year,k.aom,label="Assault on Minor Girls")
plt.plot(k.year,k.dv,label="Domestic Violence")
plt.plot(k.year,k.wt,label="Trafficking of Women")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.title("Crime distribution in Uttar Pradesh(2001-2021)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
gp=df.groupby("year")[["rape","k&a","dd","aow","aom","dv","wt"]].sum()
plt.figure(figsize=(10, 6))                     # it specifies the size of chart
sns.heatmap(gp,cmap="Reds",linewidths=0.5)
plt.xlabel("Crimes")
plt.ylabel("Year")
plt.title("Crimes on National Level")
plt.show()
```

```python
gp=df.groupby("year")["rape"].mean()
print(gp.sort_values(ascending=False))
plt.figure(figsize=(12,6))
sns.barplot(x=gp.index,y=gp.values,color="r")
plt.title("Average rape cases per year in India")
plt.xlabel("Year")
plt.ylabel("Average rape cases")
plt.show()                                       # Most rape cases reported in 2016 and least rape cases reported in 2003.
```
